```python
#lets read JoiningData.xlsx 

table1 = pd.read_excel('JoiningData.xlsx',
                   sheet_name = 0            #for sheet 1 we can wrute 0 and for sheet 2 we can write 1 and so on; OR write name of the sheet directly like 'Customers'
                   )
table1.head(5)
```

```python
table2 = pd.read_excel('JoiningData.xlsx',
                   sheet_name = 'Orders'                #for sheet 2 write name of the sheet directly like 'Orders'
                   )
table2.head(5)
```

```python
#here we have CustomerID common in both sheets which will be used in joining the data using merge()

combined1 = pd.merge(table1,
                     table2,
                     how = 'inner',             #gets only common data in both the tables
                     on = 'CustomerID')
combined1.head(5)
```

```python
combined1.shape
```

```python
combined2 = pd.merge(table1,table2,
                     how = 'left',             #gets only common data in both the tables and all data from table 1
                     on = 'CustomerID')
combined2.head(5)
```

```python
combined2.shape
```

Similarly we can do 'right' & 'outer' joins as well
